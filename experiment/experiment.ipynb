```python
import os
import pandas as pd
def get_done_set(out_file):
    """Đọc file Excel {alg}.xlsx, trả về set các dataset đã có."""
    if not os.path.exists(out_file):
        return set()
    try:
        df = pd.read_excel(out_file, engine="openpyxl")
    except Exception:
        df = pd.read_excel(out_file)
    if "dataset" not in df.columns:
        return set()
    return set(df["dataset"].astype(str).tolist())
```
